Dabei entspricht eine Periode der Abfolge eines schwarzen und eines weißen Streifens. Der Parameter `Frequenz` hat für die Struktur `streifen` keinen Einfluss.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import Layout, interact, FloatSlider, IntSlider, Dropdown
from scipy import fftpack, ndimage
from scipy.ndimage import gaussian_filter

%matplotlib inline
plt.rcParams['figure.figsize'] = [10, 5]

def _rand_dirichlet(img, rotation, order, cval=0, mode="constant"):
    return ndimage.rotate(
            img,
            rotation,
            reshape=False,
            order=order,
            mode=mode,
            cval=cval)

def _orientierter_sinus(randbedingung, size, rotation, frequenz):
    x = np.arange(size)
    X, Y = np.meshgrid(x, x)
    theta = np.pi / 4  
    
    if randbedingung == 'Periodisch':
        theta += np.deg2rad(rotation)   
     
        # diskrete Frequenzindizes -> exakt periodisch. Achtung: Rundungsfehler für beliebige Winkel möglich
        # Für periodische Randbedingungen werden ganzzahlige Frequenzindizes benötigt.
        # Durch round(...) wird der gewünschte Winkel nur approximiert. 
        # peaks liegen exakt bei (±nx,±ny), nicht unbedingt exakt beim eingegeben Winkel
        nx = round(frequenz*np.cos(theta))
        ny = round(frequenz*np.sin(theta))
          
        return np.sin(2*np.pi*(nx*X + ny*Y)/size)
    else: 
        # kontinuierliche Richtung -> nicht zwingend patch-periodisch
        omega = 2 * np.pi * frequenz / size # Frequenz = Anzahl Perioden im Bild, ω=2π k/N
        img = np.sin(omega * (np.cos(theta)*X + np.sin(theta)*Y))
        img = _rand_dirichlet(img, rotation, order=1) # endlichen Bildausschnitt rotieren
        return img
               
def _streifen(randbedingung, size, rotation, streifenbreite):
    x = np.arange(size)
    X, Y = np.meshgrid(x, x)
    basis_theta = 0 # winkel
    theta = basis_theta + np.deg2rad(rotation)
    
    
    if randbedingung == "Periodisch":
        m=10
        frequenz = size / (2 * streifenbreite)
        periode = 2 * streifenbreite
 
        nx = round(m * np.cos(theta))
        ny = round(m * np.sin(theta))
        
        g = np.gcd(nx, ny)
        nx //= g
        ny //= g

        # Spezialfälle vermeiden
        if nx == 0 and ny == 0:
               nx = 1
                
        # Warnung, falls es nicht exakt in das Bild passt
        # if size % periode != 0:
        #    print("Warnung: Streifenbreite passt nicht exakt periodisch zur Bildgröße.")
        
        # effektiver Winkel
        theta_eff = np.arctan2(ny, nx)
        
        # Winkeldifferenz in [-pi, pi]
        delta = np.arctan2(
            np.sin(theta_eff - theta),
            np.cos(theta_eff - theta)
        )

        delta_deg = np.rad2deg(abs(delta))

        if delta_deg > 0.5:
            print(
                f"Warnung: gewünschter Winkel {rotation:.1f}°, "
                f"realisiert {np.rad2deg(theta_eff):.2f}° "
                f"(Abweichung {delta_deg:.2f}°)"
            )
         
        phase = nx * X + ny * Y 

        img = ((phase % periode) < streifenbreite).astype(float)
        # img = 2*img - 1 # Peak im Ursprung
    else:
        # kontinuierliche Richtung
        phase = np.cos(theta) * X + np.sin(theta) * Y
        img = ((phase // streifenbreite) % 2).astype(float)
        img = _rand_dirichlet(img, rotation, order=3) # endlichen Bildausschnitt rotieren
        
    return img

def _rechteck(randbedingung, size, rotation):
    x = np.arange(size)
    X, Y = np.meshgrid(x, x)

    cx = size / 2
    cy = size / 2

    theta = np.deg2rad(rotation)
    
    # Koordinaten relativ zum Zentrum
    Xc = X - cx
    Yc = Y - cy
     # inverse Rotation der Koordinaten
    Xr =  np.cos(theta) * Xc + np.sin(theta) * Yc
    Yr = -np.sin(theta) * Xc + np.cos(theta) * Yc

    w = size / 3
    h = size / 3

    if randbedingung == "Periodisch":
        # torische Koordinaten: Abstand periodisch interpretieren
        Xc = ((X - cx + size/2) % size) - size/2
        Yc = ((Y - cy + size/2) % size) - size/2

        Xr =  np.cos(theta) * Xc + np.sin(theta) * Yc
        Yr = -np.sin(theta) * Xc + np.cos(theta) * Yc

        img = ((np.abs(Xr) < w/2) & (np.abs(Yr) < h/2)).astype(float)

    else:
        img = np.zeros((size, size))
        img[size//3:2*size//3, size//3:2*size//3] = 1
        img = _rand_dirichlet(img, rotation, order=3, cval=0.5) # endlichen Bildausschnitt rotieren

    return img

def _gitter(randbedingung, size, rotation, frequenz):
    x = np.arange(size)
    X, Y = np.meshgrid(x, x)

    if randbedingung == "Periodisch":
        theta = np.deg2rad(rotation)

        nx = round(frequenz * np.cos(theta))
        ny = round(frequenz * np.sin(theta))

        img = (
            np.sin(2*np.pi*(nx*X + ny*Y)/size)
            + np.sin(2*np.pi*(-ny*X + nx*Y)/size)
        )

    else:
        # zunächst unrotiert
        img = (np.sin(2*np.pi*frequenz*X/size) + np.sin(2*np.pi*frequenz*Y/size))

            # endlichen Ausschnitt rotieren -> sichtbare Ränder
        img = _rand_dirichlet(img, rotation, order=1)
    return img

def _punkt(randbedingung, size, rotation, punktgroesse=3):
    img = np.zeros((size, size))

    c = size // 2
    r = punktgroesse // 2

    img[c-r:c+r+1, c-r:c+r+1] = 1
    
    if randbedingung == "Periodisch":
        img = _rand_dirichlet(img, rotation, order=1, mode="wrap")
       

    else:
        img = _rand_dirichlet(img, rotation, order=0, mode="constant", cval=0.5)
    return img
    

def fourier_pairs(muster='Streifen', randbedingung='dirichlet', streifenbreite = 1, size=128, rotation=0, frequenz=10,):

    #-----------------------
    
    # Muster-Erzeugung
    if muster == 'Punkt':
        # Wir machen den Punkt etwas größer (3x3), damit er mehr "Energie" hat
        img = _punkt(randbedingung, size, rotation, punktgroesse=7)
        
    elif muster  == 'Streifen':
        img = _streifen(randbedingung, size, rotation, streifenbreite)

    elif muster == 'orientierter Sinus': 
        img = _orientierter_sinus(randbedingung, size, rotation, frequenz)
            
    elif muster == 'Rechteck':
        img = _rechteck(randbedingung, size, rotation)
        
    elif muster == 'Gitter':
        img = _gitter(randbedingung, size, rotation, frequenz)
    
    # Fourier Transformation
    f_shift = np.fft.fftshift(np.fft.fft2(img))
   
    # Skalierung verbessern: 
    # Wir normalisieren die Magnitude, damit auch schwache Signale (wie beim Punkt) sichtbar werden
    magnitude = np.abs(f_shift)
    magnitude = np.log(magnitude + 0.00001) # Log-Skala für Dynamikumfang

        
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(11, 5))
    
    ax1.imshow(img, cmap='gray')
    ax1.set_title(f"Ortsraum (Bild)\n{muster} bei {rotation}°", fontsize=12)
    ax1.axis('off')
    
    # Im Spektrum nutzen wir eine kontrastreiche Map
    im2 = ax2.imshow(magnitude, cmap='magma')
    # im2 = ax2.imshow(magnitude, cmap='gray')
    ax2.set_title("Frequenzraum (Spektrum - Log-Skala)", fontsize=12)
    ax2.axis('off')
    
    plt.colorbar(im2, ax=ax2, fraction=0.046, pad=0.04)
    plt.tight_layout()
    plt.show()

interact(fourier_pairs, 
         muster=['orientierter Sinus', 'Streifen', 'Gitter', 'Rechteck', 'Punkt'],
         rotation=IntSlider(min=-180, max=180, step=5, value=0, description='Rotation:', layout=Layout(width='1000px'), continuous_update=False),
         size=IntSlider(min=100, max=1000, step=1, value=128, description='Bildgröße:', layout=Layout(width='1000px'), continuous_update=False),
         streifenbreite=IntSlider(min=1, max=50, step=1, value=2, description='Streifenbreite:', layout=Layout(width='1000px'), style={'description_width': 'initial'}, continuous_update=False),
         frequenz=FloatSlider(min=1, max=50, step=1, value=5, description='Frequenz:', layout=Layout(width='1000px'),  continuous_update=False),
         randbedingung=['Dirichlet', 'Periodisch']
        )
```

```python
def plot_nyquist_demo(schwingungen=5, abtastpunkte=12):
    plt.figure(figsize=(12, 5))
    
    # 1. Kontinuierliches Originalsignal
    x_fine = np.linspace(0, 1, 1000, endpoint=False)
    y_fine = np.cos(2 * np.pi * schwingungen * x_fine)
    
    # 2. Diskrete Abtastung
    x_sample = np.linspace(0, 1, abtastpunkte, endpoint=False)
    y_sample = np.cos(2 * np.pi * schwingungen * x_sample)
    
    # 3. Fourier-/trigonometrische Rekonstruktion aus den Abtastwerten
    N = abtastpunkte
    coeffs = np.fft.fft(y_sample) / N
    freqs = np.fft.fftfreq(N, d=1/N)

    y_recon = np.zeros_like(x_fine, dtype=complex)
    for c, f in zip(coeffs, freqs):
        y_recon += c * np.exp(2j * np.pi * f * x_fine)

    y_recon = y_recon.real
    
    # Darstellung
    plt.plot(
        x_fine, y_fine,
        label=f'Originalsignal: {schwingungen} Schwingungen',
        color='blue', alpha=0.3, lw=2
    )
    
    plt.stem(
        x_sample, y_sample,
        linefmt='red', markerfmt='ro',
        basefmt=' ',
        label='Abtastwerte'
    )
    
    plt.plot(
        x_fine, y_recon,
        'r--', alpha=0.8,
        label='Fourier-Rekonstruktion aus den Abtastwerten'
    )
    
    plt.xlabel("Position im Bild (normiert von 0 bis 1)")
    plt.ylabel("Intensität / Amplitude")
    plt.legend(loc='upper right', frameon=True, shadow=True)
    plt.grid(True, linestyle=':', alpha=0.7)
    plt.ylim(-1.2, 1.6)
    plt.show()
 # Interaktive Steuerung
interact(plot_nyquist_demo, 
         schwingungen=IntSlider(min=1, max=15, step=1, value=6, description=f'Frequenz:', continuous_update=False,  layout=Layout(width='1000px'),), 
         abtastpunkte=IntSlider(min=2, max=40, step=1, value=8, description='Abtastpunkte:', style={'description_width': 'initial'}, layout=Layout(width='1000px'), 
                                continuous_update=False))
```

**Die Herleitung des Laplacian of Gaussian (LoG)**

In der Vorlesung haben wir die Formel für den LoG-Filter direkt kennengelernt. Um zu verstehen, woher diese Formel eigentlich kommt, leiten wir sie hier analytisch her.

Gegeben ist die (nicht normierte) 2D-Gauß-Funktion:
$$G_\sigma(x, y) = e^{-\frac{x^2+y^2}{2\sigma^2}}$$

Der Laplace-Operator $\nabla^2$ ist die Summe der ungemischten zweiten partiellen Ableitungen:
$$\nabla^2 = \frac{\partial^2}{\partial x^2} + \frac{\partial^2}{\partial y^2}$$
