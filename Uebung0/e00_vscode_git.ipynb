## Einführung von Visual Studio Code

Der webbasierte Editor von `jupyter lab` ist vermutlich der einfachste Einstieg für die Entwicklung interaktiver Projekte. Für alle, die eine leistungsfähigere Entwicklungsumgebung bevorzugen, stellen wir außerdem Visual Studio Code vor.  [Visual Studio Code](https://code.visualstudio.com/Download) ist ein plattformunabhängiger Open-Source-Codeeditor, der kostenlos für Windows, macOS und Linux verfügbar ist. Verwechseln Sie ihn nicht mit *Visual Studio*, der kommerziellen Entwicklungsumgebung von Microsoft für Windows.

Zur Installation von Visual Studio Code können Sie entweder die Installationsdatei über die offizielle Website herunterladen und ausführen oder das Programm direkt über den  [Anaconda Navigator](https://docs.anaconda.com/anaconda/navigator/) installieren. Nach erfolgreicher Installation können Sie Visual Studio Code entweder über das Programmsymbol starten oder durch Eingabe von `code` im Terminal Ihres Betriebssystems öffnen.



Um komfortabel mit Python und Jupyter Notebooks zu arbeiten, benötigen wir die offizielle Python](https://marketplace.visualstudio.com/items?itemName=ms-python.python) für Visual Studio Code. Diese ergänzt Funktionen wie Codevervollständigung, Debugging und Linting. Sie können die Erweiterung über den integrierten Extension Manager installieren. Klicken Sie dazu auf das Erweiterungssymbol in der linken Seitenleiste oder drücken Sie `Strg+Shift+X`. Dadurch öffnet sich das Erweiterungsfenster. Geben Sie „Python“ ein und wählen Sie den obersten Treffer ([Python Erweiterung](https://marketplace.visualstudio.com/items?itemName=ms-python.python)). Klicken Sie anschließend im rechten Bereich auf „Installieren“. Nach einem Neustart von Visual Studio Code ist die Installation abgeschlossen.

Öffnen Sie danach Visual Studio Code, wählen Sie Datei → Ordner öffnen und navigieren Sie zu einem Verzeichnis, in dem Sie Ihre Python-Dateien speichern möchten. Erstellen Sie eine neue Datei und speichern Sie diese als `test.py`.

Sehen wir uns nun an, wie sich Jupyter direkt innerhalb von VS Code nutzen lässt. Kopieren Sie dazu den folgenden Code in Ihre Python-Datei (die letzte Zeile enthält absichtlich einen Fehler):

````python
#%%
x = 'Hello World'

#%%
print(x)

#%% 
x:= 42
````

Wie Sie oben sehen, haben wir einige Zeilen mit `#%%` eingefügt, also das Kommentarzeichen `#` gefolgt von `%%`. Für Visual Studio Code markiert dies den Beginn einer ausführbaren jupyter-Zelle. Über diesen Zeilen erscheint dann ein sogenanntes Code Lens mit der Option „Run Cell“. Der Code innerhalb einer solchen jupyter-Zelle (begrenzt durch die` #%%`-Markierungen) kann anschließend durch Klicken auf „Run Cell“ ausgeführt werden.

Probieren Sie dies zunächst mit der ersten Zelle aus. Beim ersten Ausführen startet Visual Studio Code automatisch einen `jupyter`-Server mit einem neuen python-Kernel. Sobald dieser bereit ist, öffnet sich auf der rechten Seite das Fenster „Python Interactive“, in dem sowohl der eingegebene Code als auch die entsprechenden Ausgaben angezeigt werden. Wenn Sie anschließend die zweite Zelle ausführen, werden sowohl eine Eingabe- als auch eine Ausgabenzelle ergänzt. Falls Sie den jupyter-Kernel nicht verwenden möchten, können Sie die gesamte Datei auch direkt mit dem normalen Python-Interpreter ausführen. Drücken Sie dazu einfach F5 oder wählen Sie „Debug → Start Debugging“. Beachten Sie, dass die Zellmarkierungen (z.B. `#%%`) dabei ignoriert werden, da sie vom `python`-Interpreter als Kommentare behandelt werden.

Wie Sie in der dritten Zelle sehen können, hebt Visual Studio Code Syntaxfehler (und in gewissem Umfang auch Typfehler) bereits während der Eingabe automatisch hervor. Darüber hinaus zeigt Visual Studio Code automatisch Dokumentation an, die aus den Docstrings von Python-Klassen und -Funktionen extrahiert wird. Sie können dies beispielsweise ausprobieren, indem Sie `print`( eingeben – daraufhin werden Ihnen die Parameter der `print`-Funktion angezeigt. Diese intelligente Unterstützung können Sie auch manuell mit Strg+Leertaste aufrufen.

Abschließend bietet Visual Studio Code eine integrierte Unterstützung für `git`-Repositories (s. E00-git.ipnyb). Sobald Sie eine Datei oder ein Verzeichnis öffnen, das sich innerhalb eines `git`-Repositories befindet, wird in der linken unteren Ecke der Statusleiste ein Branch-Symbol (z. B. „master“) angezeigt. Daneben befindet sich ein Synchronisationssymbol. Durch Klicken auf dieses Symbol können Sie automatisch Änderungen aus dem entfernten Repository herunterladen (pull) und lokale Commits hochladen (push).  Über den `git`-Bereich in der linken Navigationsleiste (drittes Symbol von oben) gelangen Sie zu einer Übersicht Ihrer lokalen Änderungen und nicht verfolgten Dateien. Dort können Sie gezielt Änderungen auswählen, die Sie `committen` möchten, eine Commit-Nachricht eingeben und mit Strg+Enter einen lokalen Commit erstellen. Zusätzlich haben Sie die Möglichkeit, Änderungen manuell zu pushen oder pullen.



## Verwendung des verteilten Versionsverwaltungssysmtes `git`

Auch wenn Sie den gesamten Code und die benötigten Daten für die Übungen über StudIP herunterladen können, empfehlen wir die Verwendung des verteilten Versionsverwaltungssystems `git` zur Versionierung und Organisation Ihres Codes. `git` ist generell für alle Dateien und Verzeichnisse zu empfehlen, bei denen Versionierung sinnvoll ist – selbst wenn Sie alleine auf Ihrem lokalen Rechner arbeiten. Es kann beispielsweise sehr hilfreich sein, wenn Sie versehentlich eine Datei oder einen Abschnitt in einem Bericht löschen, da Sie jederzeit zwischen verschiedenen Versionen Ihrer Arbeit wechseln können.

Im Folgenden geben wir eine kurze Einführung in `git`. `git` ist ein sehr leistungsfähiges Werkzeug: Während grundlegende Funktionen einfach zu erlernen sind, erfordern fortgeschrittene Anwendungen – insbesondere in kollaborativen Umgebungen – etwas Übung. `git` wird heute breit eingesetzt, etwa in der Softwareentwicklung, in datengetriebenen Projekten sowie allgemein bei der kollaborativen Bearbeitung von Dateien. Ältere Versionsverwaltungssysteme wie CVS oder SVN wurden dadurch weitgehend abgelöst.

Stellen wir zunächst sicher, dass `git` auf Ihrem System installiert ist. Einige Linux-Distributionen liefern `git` bereits mit oder ermöglichen die Installation über die offiziellen Paketquellen (z. B. `apt` unter Ubuntu). Für macOS und Windows können Sie die `git`-Kommandozeilenwerkzeuge über die [offizielle Website](https://git-scm.com/download/) herunterladen. Vorausgesetzt, git ist korrekt installiert, können Sie ein entferntes Repository klonen, indem Sie den folgenden Befehl im Terminal ausführen:
``` git clone ```
Beachten Sie, dass die oben erwähnte `git`-Distribution ein spezielles Terminal namens Git Bash mitliefert, das die Arbeit mit `git` besonders erleichtert.

Die grundlegende Einheit in `git` (sowie in CVS und SVN) ist ein sogenanntes Repository. Dieses kann man sich als ein versioniertes Verzeichnis vorstellen, das Dateien und Unterverzeichnisse enthält. Wie bei CVS oder SVN benötigen wir zunächst eine lokale Kopie eines Repositories. In diesem lokalen Repository können wir Dateien bearbeiten. Sobald wir einen Zustand erreicht haben, den wir als neue Version speichern oder mit anderen teilen möchten, führen wir einen commit aus. Dabei werden die vorgenommenen Änderungen dauerhaft gespeichert, während frühere Versionen der Dateien erhalten bleiben. Falls Sie mit CVS oder SVN vertraut sind, wissen Sie, dass diese Systeme auf einem zentralen Server basieren. Ein `commit` überträgt dort die lokalen Änderungen direkt auf das zentrale Repository. Ein wesentlicher Unterschied bei `git` besteht darin, dass Repositories dezentral organisiert sind. Für einen `commit` ist kein Server erforderlich; stattdessen speichert jedes lokale `git`-Repository die vollständige Historie aller Änderungen eigenständig. Die Zusammenarbeit mit anderen ist dennoch problemlos möglich, indem man lokale Kopien (Klone) von Repositories erstellt. Änderungen können anschließend zwischen Repositories ausgetauscht werden, indem man entfernte Änderungen in das lokale Repository übernimmt (pull) oder eigene Änderungen an ein entferntes Repository überträgt (push).

Wenn mehrere Personen zusammenarbeiten, kann ein entferntes Repository auf einem Server gehostet werden, das dann ähnlich wie ein zentraler Server in CVS oder SVN fungiert. Wichtig ist jedoch, dass alle Klone dieses Repositories weiterhin die vollständige Versionshistorie enthalten. Das bedeutet, dass selbst bei eine m späteren Löschen des zentralen Repositories keine Informationen verloren gehen. Eine verbreitete Möglichkeit, solche `git`-Repositories zentral zu hosten, bieten Plattformen wie `GitHub` oder institutionelle `GitLab`-Server, wie der [GitLab-Server der Uni Osnabrueck](https://gitlab.uni-osnabrueck.de).

Legen wir nun praktisch los und arbeiten mit `git`. Zunächst erstellen wir ein lokales Repository in unserem Dateisystem. Dazu öffnen wir ein `git`-Terminal, wechseln in das Verzeichnis, das die zu versionierenden Dateien enthalten soll, und führen folgenden Befehl aus:

```
> git init
```
Sobald wir uns in einem Unterverzeichnis eines `git`-Repositories befinden, können wir folgenden Befehl ausführen:

```
> git status
```

um den aktuellen Status des Repositories zu überprüfen. Dabei wird angezeigt, ob (i) neu erstellte, nicht verfolgte Dateien (untracked files) vorhanden sind oder (ii) Änderungen an bereits verfolgten Dateien vorgenommen wurden, die noch nicht committet wurden. Probieren wir dies aus, indem wir eine neue Textdatei `file1.txt` erstellen und anschließend erneut `git status` ausführen. Wir sehen nun, dass `git` eine untracked Datei erkannt hat, also eine Datei, die lokal existiert, aber noch nicht dem Repository hinzugefügt wurde.
Standardmäßig sind Dateien zunächst untracked, solange sie nicht explizit hinzugefügt werden. Das ist sinnvoll, da häufig temporäre Dateien entstehen, die nicht versioniert werden sollen.
Mit dem Befehl `git add` können wir eine Datei explizit zum Repository hinzufügen, also beispielsweise:


```
> git add file1.txt
```

Mit diesem Befehl teilen Sie `git` im Wesentlichen mit, dass eine Datei künftig versioniert werden soll, ohne sie jedoch bereits zu `committen`. Dies können wir anschließend durch folgenden Befehl tun:

```
> git commit -am 'added a file'
```
Die Option `-a` sorgt dafür, dass alle Änderungen an bereits verfolgten Dateien automatisch committet werden, berücksichtigt jedoch keine neuen Dateien, die nicht zuvor mit `git add` zum Tracking hinzugefügt wurden.
Mit der Option `-m `können Sie eine Commit-Nachricht direkt in der Kommandozeile angeben. Wenn Sie nun den Status erneut überprüfen, indem Sie wieder


```
> git status
```

aufrufen, werden Sie sehen, dass nun Änderungen vorhanden sind, die noch nicht committet wurden. Außerdem zeigt der folgende Befehl:

```
> git log
```
die Historie des Repositories an, die in unserem Fall aus genau einem Commit besteht. Nehmen wir nun eine Änderung an der Datei `file1.txt` vor, committen diese und überprüfen anschließend erneut den Log.

```
> git commit -am 'changed file'
> git log
```
Es ist nicht überraschend, dass der Log nun zwei Einträge enthält: einen, der den Zustand der Datei vor der Änderung beschreibt, und einen, der den Zustand nach der Änderung repräsentiert.
Mit den Befehlen `git show` und `git checkout` können wir uns die Änderungen eines bestimmten Commits anzeigen lassen und bequem zwischen verschiedenen Versionen wechseln. Zur eindeutigen Identifikation eines Commits verwendet `git` sogenannte Hash-Werte, die als lange Folge von Buchstaben und Zahlen dargestellt werden. Diese Hash-Werte sind beispielsweise in Einträgen im Log wie folgt zu sehen:

```
commit 754592e1397ff7a5e4dbf568537419d254d081f9 (HEAD -> master)
Author: vanbetteray <antonia.vanbetteray@uos.de>
Date:   Tue Apr 7 22:24:34 2026 +0200

    added a file
```

Keine Sorge, Sie müssen nicht den gesamten Hash-Wert eingeben, um den Inhalt eines Commits anzuzeigen. Es genügt, die ersten Zeichen anzugeben, die den Commit innerhalb Ihres Repositories eindeutig identifizieren:

```
> git show 754592
```
Dabei werden Zeilen, die im Vergleich zur vorherigen Version gelöscht wurden, rot dargestellt, und neu hinzugefügte Zeilen grün. Wenn wir zur vorherigen Version der Datei wechseln möchten, können wir diese einfach mit dem Befehl `git checkout` auschecken, indem wir den Hash des gewünschten Commits angeben:


```
> git checkout 5126
```
Sie werden feststellen, dass sich die Datei nun im Zustand vor dem letzten Commit befindet. Um wieder zur neuesten Version im Hauptzweig (master) unseres Repositories zurückzukehren, können wir einfach folgenden Befehl ausführen:

```
> git checkout master
```
und Sie werden feststellen, dass sich die Datei wieder im neuesten Zustand befindet. Diese einfachen Befehle können beim Entwickeln von Code oder beim Schreiben von Berichten und Abschlussarbeiten sehr hilfreich sein.
Zudem ermöglicht git, die gesamte Historie eines Projekts zu archivieren, indem einfach der Ordner gesichert wird, der das Repository enthält. Die Repository-Metadaten, einschließlich früherer Versionen von Dateien, werden automatisch im versteckten Verzeichnis `.git` gespeichert.
