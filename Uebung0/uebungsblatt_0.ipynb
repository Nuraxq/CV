## Aufgabe 1: Helligkeitsanpassung

In einem Graustufenbild repräsentiert jeder Pixel einen Helligkeitswert im Bereich $[0, 255]$,
wobei $0$ Schwarz und $255$ Weiß entspricht.

Wir setzen im folgenden $\alpha = 1$ und erhalten damit:
$$p_{\text{out}} = \alpha \cdot p_{\text{in}} + \delta = p_{\text{in}} + \delta$$

wobei $p_{\text{in}}$ der Helligkeitswert des Eingabepixels und $p_{\text{out}}$ der des Ausgabepixels ist.

**Aufgabe 1.1:** Gegeben sei ein Helligkeitswert $p_{\text{in}} = 240$ und ein Aufhellungswert $\delta = 30$.
Berechnen Sie $p_{\text{out}}$ und geben Sie ihn aus. Was fällt auf?



````python
p_in  = 240
delta = 30

# TODO: Berechnen Sie p_out und geben Sie ihn aus
p_out = p_in + delta
print(f"p_out = {p_out}")
````

## Aufgabe 2: Clamping mit `if`

Um Overflow und Underflow zu verhindern, müssen wir den Ausgabewert auf $[0, 255]$
begrenzen. Diesen Vorgang nennt man **Clamping**.

*Hinweis:* Es gibt drei Fälle: zu klein, zu groß, oder im gültigen Bereich.


````python
def clamp_if(x: float) -> int:
    """
    Begrenzt x auf den Bereich [0, 255] und gibt das Ergebnis als int zurück.

    Parameters:
        x: Eingabewert.

    Returns:
        Geclampter Wert als int in [0, 255].
    """
    # TODO
    if(x > 255):
        x = 255
    elif (x < 0):
        x = 0
    return int(x)

    pass
````

## Aufgabe 3: Clamping mit `min` und `max`

Die `if`-Variante funktioniert, ist aber verhältnismäßig lang.
Python bietet die eingebauten Funktionen `min(a, b)` und `max(a, b)`,
die den kleineren bzw. den größeren zweier Werte zurückgeben:

```python
max(5, 6)   # 6  – verhindert Underflow
min(5, 6)   # 5  – verhindert Overflow
```

**Aufgabe 3.1:** Implementieren Sie `clamp(x)` in **einer Zeile** ohne `if`,
indem Sie `min` und `max` kombinieren.


````python
def clamp(x: float) -> int:
    """
    Begrenzt x auf den Bereich [0, 255] ohne if-Statements.

    Parameters:
        x: Eingabewert.
    Returns:
        Geclampter Wert als int in [0, 255].
    """
    # TODO: eine Zeile, kein if

    return max(min(255,x),0)

    pass
````

````python
# --- Test 3.1 ---
assert clamp(270)  == 255
assert clamp(-10)  == 0
assert clamp(128)  == 128
assert clamp(0)    == 0
assert clamp(255)  == 255
assert isinstance(clamp(100), int), "Rückgabetyp muss int sein"
print("✅ clamp korrekt!")
````

```python
# for-Schleife:
```

## Aufgabe 6: Lambda-Ausdrücke und Funktionen als Parameter

In Python gilt *everything is an object* – auch Funktionen.
Man kann sie daher als Argument an andere Funktionen übergeben.
Für kurze, anonyme Funktionen bieten sich **Lambda-Ausdrücke** an.

**Syntax:**

```python
lambda parameter1, parameter2, ...: ausdruck
```

**Vergleich mit einer normalen Funktion:**

```python
# Normale Funktion:
def verdoppeln(x):
    return x * 2

# Äquivalentes Lambda:
verdoppeln_lambda = lambda x: x * 2

print(verdoppeln_lambda(5))  # 10
```

**Aufgabe 6.1:** Ergänzen Sie die Signatur von `brightness_adjust_variable` um
einen Parameter `my_func: Callable`, der die Clamp-Funktion von außen übergeben bekommt.

und einmal in der `min`/`max`-Variante auf und überprüfen Sie, dass beide Aufrufe
dasselbe Ergebnis liefern.



````python
from typing import Callable

def brightness_adjust_variable(bild: list[list[int]], delta: int, my_func: Callable) -> list[list[int]]:
    """
    Passt die Helligkeit eines Bildes an, wobei die Clamp-Funktion als Parameter übergeben wird.

    Parameters:
        bild:    Verschachtelte Liste von Pixelwerten [0, 255].
        delta:   Verschiebung (positiv = heller, negativ = dunkler).
        my_func: Funktion, die einen einzelnen Pixelwert auf [0, 255] begrenzt.

    Returns:
        Neues Bild mit angepassten Pixelwerten.
    """
    # TODO: Comprehension of Comprehensions mit my_func statt clamp

    newList = [[my_func(x,delta) for x in zeile] for zeile in bild]

    return newList

    pass
````

````python
# TODO: Definieren Sie die beiden Lambda-Ausdrücke
clamp_if_lambda      = lambda x,delta: 255 if x+delta > 255 else 0 if x+delta < 0 else x+delta
clamp_min_max_lambda = lambda x,delta: max(min(255,x+delta),0)

# --- Test 6.1 ---
bild_test = [[0, 100, 200, 240], [50, 150, 250, 10]]

result_if     = brightness_adjust_variable(bild_test, 30, clamp_if_lambda)
result_minmax = brightness_adjust_variable(bild_test, 30, clamp_min_max_lambda)
result_ref    = brightness_adjust(bild_test, 30)

assert result_if == result_ref,     f"clamp_if_lambda weicht ab: {result_if}"
assert result_minmax == result_ref, f"clamp_min_max_lambda weicht ab: {result_minmax}"
print("✅ brightness_adjust_variable korrekt!")
````

## Aufgabe 7: Slicing – Teilbereiche aus Arrays auslesen

Bevor wir das ganze Bild auf einmal verarbeiten, schauen wir uns an, wie man
mit NumPy gezielt **Teile** eines Arrays liest – ohne Schleifen.

### Warum nicht einfach Python-Listen?

Versuchen wir zunächst, was passiert wenn wir eine Python-Liste mit einem Skalar multiplizieren:

```python
zeile = [10, 20, 30]
zeile * 2   # ???
```

Statt jeden Wert zu verdoppeln, wiederholt Python die gesamte Liste: `[10, 20, 30, 10, 20, 30]`.
Das ist für Bildverarbeitung **unbrauchbar** – man müsste wieder eine `for`-Schleife schreiben.

NumPy-Arrays verhalten sich dagegen wie Vektoren aus der Mathematik:
```python
a = np.array([10, 20, 30])
a * 2   # [20, 40, 60]  ✅  – jedes Element wird verdoppelt
a + 5   # [15, 25, 35]  ✅  – Broadcasting
```

### Slicing-Syntax für 2D-Arrays

Ein Bild ist ein 2D-Array mit Form `(Zeilen, Spalten)`. Mit `[zeile_idx, spalten_idx]`
greift man auf einzelne Elemente oder ganze Bereiche zu:

```python
bild = np.array([
    [100, 200,  50, 180],   # Zeile 0
    [ 30,  80, 210,  60],   # Zeile 1
    [  5, 120,  90, 240],   # Zeile 2
])

bild[1, 2]        # Einzelner Pixel: 210
bild[0, :]        # Ganze Zeile 0:   [100, 200, 50, 180]
bild[:, 1]        # Ganze Spalte 1:  [200, 80, 120]
bild[0:2, 1:3]    # Teilbereich:     [[200, 50], [80, 210]]
bild[-1, :]       # Letzte Zeile:    [5, 120, 90, 240]
```

Das `:` bedeutet „alle" – `bild[0, :]` heißt „Zeile 0, alle Spalten".
Ein Bereich `a:b` wählt die Indizes `a, a+1, ..., b-1` (Ende ist **exklusiv**, wie in C++).

## Aufgabe 9: Kontrastskalierung mit NumPy ($\alpha \cdot p_{\text{in}} + \delta$)

In Aufgabe 1 haben wir nur den Fall $\alpha = 1$ betrachtet.
Die allgemeine Formel lautet:

$$p_{\text{out}} = \alpha \cdot p_{\text{in}} + \delta$$

Werte von $\alpha > 1$ erhöhen den **Kontrast** (helle Pixel werden heller, dunkle noch dunkler),
$\alpha < 1$ reduziert ihn.
In den bisherigen Aufgaben haben die Helligkeitswerte ausschließlich ganzzahlige Werte (`int`) angenommen. Durch die Multiplikation mit  $\alpha$ können jedoch Gleitkommawerte (`float`) entstehen, sodass die Helligkeitswerte auch nicht mehr notwendigerweise ganzzahlig sind.

Mit NumPy lässt sich das direkt auf das gesamte Bild anwenden – wieder ohne Schleife.

**Aufgabe 9.1:** Implementieren Sie `contrast_adjust(bild, alpha, delta)`.



````python
def contrast_adjust(bild: np.ndarray, alpha: float, delta: int) -> np.ndarray:
    """
    Wendet die Transformation p_out = alpha * p_in + delta an und clampt auf [0, 255].

    Parameters:
        bild:  2D-Array mit Pixelwerten [0, 255].
        alpha: Kontrastfaktor (> 1: mehr Kontrast, < 1: weniger Kontrast).
        delta: Helligkeitsverschiebung (Bias).

    Returns:
        Neues Array mit geclampten Werten.
    """
    # TODO: eine Zeile – kein for, kein clamp()

    return np.clip(alpha*bild+delta,0,255)
    pass
````
