## Verwendung von `jupyter` notebooks

Nachdem alle Pakete installiert wurden, sollten Sie in der Lage sein, einen Jupyter-Server zu starten. Dabei handelt es sich um einen Prozess, der Python-Code in sogenannten Kernels ausführt und die Ergebnisse zurückliefert. 
Dieser Jupyter-Server kann lokale Python-Kernels betreiben, die im Hintergrund als eigenständige Prozesse laufen. Sie empfangen Anfragen mit Python-Code, führen diesen aus und geben die Ergebnisse zurück. Eine verbreitete – wenn auch nicht die einzige – Möglichkeit, mit solchen Kernels zu interagieren, sind die browserbasierten Oberflächen `jupyter notebook` und `jupyter lab`.

Beide Oberflächen sind sich insofern ähnlich, als sie es ermöglichen, `jupyter`-Notebooks direkt im Browser zu öffnen, zu bearbeiten und auszuführen. Ein Notebook besteht aus mehreren Eingabe- und Ausgabezellen. In den Eingabezellen wird der auszuführende Code definiert, während die zugehörigen Ausgabezellen die Ergebnisse der Berechnung anzeigen. Wichtig ist, dass diese Zellen Code in verschiedenen Programmiersprachen wie python, `julia` oder `R` enthalten können (daher der Name `jupyter`), aber auch Markdown-formatierten Text, HTML-Code oder sogar LaTeX-Formeln. Dadurch eignen sich `jupyter`-Notebooks hervorragend zur Erstellung interaktiver, ausführbarer Dokumente, die sich beispielsweise für Computer-Vision-Experimente, Auswertungen und Berichte nutzen und direkt in HTML oder LaTeX/PDF exportieren lassen.

Für diesen Kurs können Sie entweder eine lokale Umgebung in einer IDE wie VS Code \(s. e00\_vscode\_git.ipynb\) verwenden oder den Server `jupyter notebook` als auch `jupyter lab`, wobei letzterer die modernere Variante darstellt. Für die installation mit `conda` führen Sie den folgenden Befehl aus:

```
conda install -c conda-forge jupyterlab
```
Aktivieren Sie die `conda`-Umgebung mit 

```conda activate my_env```.

Um einen `jupyter lab`\-Server zu starten, navigieren Sie zunächst im Terminal in das Verzeichnis Ihres Dateisystems, in dem Sie ein `jupyter`\-Notebook erstellen oder öffnen möchten, und führen Sie anschließend den folgenden Befehl aus:

```
> jupyter lab
```

Als ersten Schritt können Sie dies im Verzeichnis ausprobieren, das dieses Notebook enthält \(d. h. die entsprechende `.ipynb`\-Datei\). Es öffnet sich ein Browserfenster, in dem Sie im linken Bereich den Dateibrowser sehen sollten. Dort wird das Notebook angezeigt – klicken Sie darauf, um es als interaktives Dokument zu öffnen. Auf ähnliche Weise können Sie auch ein neues Notebook erstellen. Durch Doppelklicken auf eine Eingabezelle können Sie den enthaltenen Text oder Code bearbeiten. Mit Shift\+Enter führen Sie die Zelle aus; dabei wird entweder formatierter Text oder die Ausgabe des zugrunde liegenden Codes angezeigt.

Probieren wir dies in der folgenden Eingabezelle aus, die `python`-Code enthält und eine einfache textuelle Ausgabe erzeugt:



````python
x = 2 * 22
print(x)
````

Sie können unterhalb der aktuellen Cursorposition eine neue Eingabezelle erstellen, indem Sie entweder auf die Schaltfläche `+`  im Menü oben links klicken oder die Taste `b` auf der Tastatur drücken. Wenn Sie eine Zelle bearbeiten, können Sie mit `Esc` in den Kommando-Modus wechseln. In diesem Modus lassen sich Zellen hinzufügen, verschieben oder löschen. Zum Löschen einer Zelle drücken Sie im Kommando-Modus zweimal D. Um den Zelltyp von `python` auf markdown zu ändern, drücken Sie im Kommando-Modus `m`. Mit `y` wechseln Sie wieder zurück zu `python`-Code.

Probieren wir dies nun mit der folgenden Markdown-Zelle aus, die eine LaTeX-Formel enthält:



**Hinweis**: Wir können auch LaTeX-Formeln verwenden: $$\int_0^\pi \sin(t) \, dt$$



Der gesamte Python-Code wird vom zugrunde liegenden Python-Kernel ausgeführt, dessen aktueller Status durch den Kreis oben rechts im Notebook-Fenster angezeigt wird. Ein nicht ausgefüllter Kreis (d. h. mit weißem Inneren) zeigt an, dass der Kernel momentan inaktiv ist. Ist der Kreis schwarz ausgefüllt, führt der Kernel gerade Berechnungen aus.

Dies können Sie beobachten, wenn Sie die folgende Zelle ausführen (Shift+Enter):

````python
s = 0
for i in range(80000000):
    s += 1
print(s)
````

````python
print(s)
````

Abschließend ist es wichtig zu beachten, dass der Python-Kernel lediglich ein einzelner Interpreter-Prozess ist, der den Code in den ausgeführten Zellen nacheinander abarbeitet. Das bedeutet, dass die Reihenfolge der Ausführung den aktuellen Zustand des Kernels bestimmt, also welche Variablen existieren und welche Werte sie besitzen.

Insbesondere bleibt der Zustand über mehrere Zellen hinweg erhalten, wie Sie im folgenden Beispiel sehen können:

````python
x = 42
````

````python
print(x)
````

Wenn Sie diese beiden Zellen in umgekehrter Reihenfolge ausführen, entsteht ein Fehler. Das wirkt zunächst trivial, kann jedoch bei komplexeren Notebooks, in denen Zellen mehrfach und in unterschiedlicher Reihenfolge ausgeführt werden, schnell unübersichtlich werden, da der aktuelle Zustand des Kernels nicht immer offensichtlich ist. Sie können den aktuellen Zustand jederzeit zurücksetzen, indem Sie den Kernel neu starten, wodurch ein neuer Interpreter-Prozess gestartet wird. Dies können Sie entweder über die Option „Restart Kernel“ im Menü Kernel tun oder indem Sie in den Kommando-Modus wechseln (`ESC` drücken) und anschließend zweimal die Taste `0` drücken. Probieren Sie dies aus und führen Sie danach die folgende Zelle aus – diese sollte einen Fehler erzeugen, da die Variable `x` im neuen Kernel noch nicht definiert wurde.


````python
print(x)
````
